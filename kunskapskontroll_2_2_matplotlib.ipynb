### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Explicit. Ett objektorienterat sätt där man skapar två objekt "Figure" och "Axes". T.ex. på detta sätt fig, ax = plt.subplots(). Sedan använder man dessa objekts inbygda metoder. 

Implicit. Då skapar man grafer direkt genom att anropa modul-funktioner i Matplotlib, t.ex. så här plt.plot()


Den explicite metoden är att rekommendera då den ger mer kontroll. Ska man dock göra något enkelt och snabbt kan man använda den implicita metoden. 

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: Det är hela ytan som visualiseringen sker. Kan innehålla flera grafer m.m. Man kan se det som en hel, tom målarduk.

Axes: En individuell graf/plot inom Figure. Varje Axes har en rubrik. 

Axis: En grafs axlar, t.ex. x- och y-axeln. Innehåller även etiketter, skala och gränser för grafen.

Artist: Varje enskilt element som plottas i en Figure. T.ex. är varje Axis en Artist och all text är Artist.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y, color='black')
ax.set_title('Graf över funktionen y=x². Explicit approach.')
ax.set_xlabel('x')
ax.set_ylabel('y=x²')
ax.grid()

plt.show()


plt.plot(x, y, color='green')
plt.title('Graf över funktionen y=x². Implicit approach.')
plt.xlabel('x')
plt.ylabel('y=x²')
plt.grid()

plt.show()
```
